```python
def fidelity():
    import pandas as pd
    import os
    import glob
    !pip install openpyxl
    !pip install xlrd

    files = glob.glob('/Users/emmanuel.olatunbosun/Downloads/9110000664-30th Apr-1st May.xlsx')

    for file in files:
        raw_file = pd.read_excel(file)

    new_file= raw_file.drop([0,1])

    new_file.columns = new_file.iloc[0]
    new_file = new_file[1:]

    new_file = new_file.drop(columns = ['SOL ID','REF NUM','TRAN ID', 'RATE','RATE CODE','Account Name','accountno','ACCT CRNCY CODE','Document Processed By','Document Authorized By'])

    new_file_1 = new_file.drop(new_file.columns[[0,1,4]], axis=1)

    new_file_1['VALUE DATE'] = new_file_1['TRAN DATE']

    new_file_1['TRAN PARTICULA'] = new_file_1['TRAN PARTICULAR'] + " " + new_file_1['TRAN RMKS']

    new_file_1 = new_file_1.drop(new_file_1.columns[[3,4]], axis=1)

    new_file_1 = new_file_1.iloc[:,[1,3,4,2,0]]

    new_file_1['TRAN DATE'] = pd.to_datetime(new_file_1['TRAN DATE'])
    new_file_1['VALUE DATE'] = pd.to_datetime(new_file_1['VALUE DATE'])

    new_file_1['TRAN DATE'] = new_file_1['TRAN DATE'].dt.strftime('%d-%b-%y')
    new_file_1['VALUE DATE'] = new_file_1['VALUE DATE'].dt.strftime('%d-%b-%y')

    new_file_1['TRAN AMT'] = new_file_1['TRAN AMT'].apply('{:,}'.format)

    new_file_1 = new_file_1.iloc[:,[0,1,2,4,3]]

    #math operations
    new_file_2 = new_file_1.copy()
    new_file_2['TRAN AMT'] = pd.to_numeric(new_file_2['TRAN AMT'].str.replace(',', ''))
    computed_fee = new_file_2.loc[new_file_2['Dr Cr'] == 'C', 'TRAN AMT'].apply(lambda x: x * 0.002 if x * 0.002 < 50 else 50).sum()
    total_credit = new_file_2.loc[new_file_2['Dr Cr'] == 'C', 'TRAN AMT'].sum()
    total_debit = new_file_2.loc[new_file_2['Dr Cr'] == 'D', 'TRAN AMT'].sum()
    debit_count = new_file_2.loc[new_file_2['Dr Cr'] == 'D', 'TRAN AMT'].count()
    difference = computed_fee - total_debit

    print('--------')
    print(f'Credit:',total_credit)
    print(f'Debit:', total_debit)
    print(f'Debit count:', debit_count)
    print (f'Fee computed:',computed_fee)
    print(f'Difference:', computed_fee - total_debit)

    return new_file_1.to_excel('sample_15.xlsx', index = False)

fidelity()
```
